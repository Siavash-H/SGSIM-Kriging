## Omni_variogram

```python
East=Data1['% Easting(km)']
# E=np.asarray(E)
North=Data1['Northing(km)']
# N=np.asarray(N)
Poros=Data1['Porosity (%)']
E1=np.asmatrix(East)
N1=np.asmatrix(North).T
Poro1=np.asmatrix(Poros)
n_points1=E1.shape[1]
ones_vec1=np.asmatrix(np.ones(n_points1))
E_mat=E1.T*ones_vec1
N_mat=N1*ones_vec1
Poro_mat=Poro1.T*ones_vec1
```

```python
dist=np.sqrt(np.power((E_mat-E_mat.T),2)+np.power((N_mat-N_mat.T),2))
sqrt_diff=np.power((Poro_mat-Poro_mat.T),2)
sqrt_diff=np.asmatrix(sqrt_diff)
```

```python
lag_size_om=1
n_lags_om=21
Evar=np.zeros(n_lags_om)
lag_tol_om=0.5*lag_size_om
for l in range(n_lags_om):
    lag=l+1
    ul_om=lag_size_om*lag+lag_tol_om
    ll_om=lag_size_om*lag-lag_tol_om
    
    index_om=np.argwhere((dist<=ul_om)&(dist>=ll_om))
    Nh=index_om.shape[0]
    
    Evar[l]=np.sum(sqrt_diff[index_om[:,0],index_om[:,1]])/(2*Nh)
```

### Plotting Omni_variogram

```python
lag_dist_om=np.arange(lag_size_om,lag_size_om*(n_lags_om+1),lag_size_om)
plt.figure(figsize=(10,8))
plt.plot(lag_dist_om[:12],Evar[:12],'-*' )
plt.title('Omni variogram')
plt.xlabel('dist')
plt.grid()
```

```python
def sph_var(h,rang,sill,nugget):
    gamma = nugget + sill*(1.5*(h/rang)-0.5*np.power((h/rang),3) )  
    index = np.argwhere(h>rang)
    if nugget < 0:
        nugget = 0
    if index.shape[1] > 1:
        gamma[index[:,0],index[:,1]] = nugget + sill
    else:
        gamma[index] = nugget + sill
    return gamma
```

```python
params1,_ = curve_fit(sph_var,lag_dist_om,Evar,p0=[3.9,0.5,0.001],bounds=(0,[7,1.5,0.35])) # po:range, sill, nugget ,bounds=(0,[80,20,2.5])
rang1 =  params1[0]
sill1 = params1[1]
nugget1 = params1[2]
print('for spherical variogram')
print('range is:',round(rang1,4))
print('sill is:',round(sill1,4))
print('nugget effect is:',round(nugget1,4))
```

# Manual fitting

```python
a = 5
c = 1.
nu = 0.04
gamma_h= sph_var(lag_dist_om,a,c,nu) # Passing the lag_dist as lag_dist_om
plt.figure(figsize=(10,8))
plt.plot(lag_dist_om[:15],Evar[:15],label='Omni')
plt.plot(lag_dist_om[:15],gamma_h[:15],label='Exponential Model') # Fitted function
plt.grid()
plt.legend()
plt.xlabel('dist')
```

## Simple Kriging with Omni_variogram

```python
for ug in range(0,len(U0_grids)):
    jj = np.ceil(U0_grids[ug]/Nx)
    ii = U0_grids[ug] - (jj - 1) * Nx
    x_cord = ((ii-1)*dx + ii * dx)/2
    y_cord = ((jj-1)*dy + jj * dy)/2
    dist_U_U0_1 = np.sqrt(np.power(X_M - x_cord,2) + np.power(Y_M - y_cord,2))
    points_within_circle = np.argwhere(dist_U_U0_1 < 4.5)
    ones_vec = np.asmatrix(np.ones(points_within_circle.shape[0]))
    X_U = X_M[points_within_circle[:,0]]
    Y_U = Y_M[points_within_circle[:,0]]
    X_mat = X_U * ones_vec
    Y_mat = Y_U * ones_vec
    dist_UU = np.sqrt(np.power(X_mat - X_mat.T,2) + np.power(Y_mat - Y_mat.T,2))
    dist_U_U0 = dist_U_U0_1[points_within_circle[:,0]]
    gamma_UU = sph_var(dist_UU,a,c,nu)
    gamma_UU0 = sph_var(dist_U_U0,a,c,nu)
    C_UU = nu +c- gamma_UU
    C_UU0 = nu + c - gamma_UU0
    cw = np.linalg.inv(C_UU) * C_UU0
    Poro_grid[int(ii)-1,int(jj)-1] =cw.T * ( poro_declust_M[points_within_circle[:,0]] - np.mean(poro_declust_M)) + np.mean(poro_declust_M)
    var_grid[int(ii)-1,int(jj)-1] = (nu + c) - (cw.T * C_UU0)
   
```

```python
f = plt.figure() 
f.set_figwidth(30)
f.set_figheight(24)
plt.imshow(Poro_grid.T,extent=[0,Lx,0,Ly],cmap='jet',vmin=12,vmax=17)
plt.title('Simple Kriging', fontsize=29)
plt.xlabel('X(Km)', fontsize=15)
plt.ylabel('Y(Km)', fontsize=15)
plt.gca().invert_yaxis()
plt.colorbar()
```

```python
X3 = Data1['% Easting(km)']
Y3 = Data1['Northing(km)']
Poro3 = Data1['Porosity (%)']

mean_poro=np.mean(Poro3)
std_poro=np.std(Poro3,ddof=1)

X3M=np.asmatrix(X3).T
Y3M=np.asmatrix(Y3).T
poro3M=np.asmatrix(Poro3).T


poro_ns=(poro3M-mean_poro)/std_poro # Data into normal scores
poro_ns200=(poro3M-mean_poro)/std_poro  # Data into normal scores
```

```python
dist_sn=np.sqrt(np.power((X3M-X3M.T),2)+np.power((Y3M-Y3M.T),2))
sqrt_diff_sn=np.power((poro_ns200-poro_ns200.T),2)
sqrt_diff_sn=np.asmatrix(sqrt_diff)
```

## Plotting Omni variogram for Standard Normalized data

```python
lag_size_om=1
n_lags_om=21
Evar_ns=np.zeros(n_lags_om)
lag_tol_om=0.5*lag_size_om
for l in range(n_lags_om):
    lag=l+1
    ul_om=lag_size_om*lag+lag_tol_om
    ll_om=lag_size_om*lag-lag_tol_om
    
    index_om=np.argwhere((dist_sn<=ul_om)&(dist_sn>=ll_om))
    Nh=index_om.shape[0]
    
    Evar_ns[l]=np.sum(sqrt_diff_sn[index_om[:,0],index_om[:,1]])/(2*Nh)
```

```python
Lx=math.ceil(np.max(X3)) #to avoid having float number in Nx, Ny
Ly=math.ceil(np.max(Y3))
dx = 0.1
dy = 0.1
Nx = np.ceil(Lx / dx)
Ny = np.ceil(Ly / dy) 
ind=np.argwhere(iy==0)
iy[ind]=1
ind=np.argwhere(ix==0)
ix[ind]=1
```

```python
loc=(iy-1)*Nx+ix
bins_points=np.arange(1,Nx*Ny+2,1)
```

```python
hist=np.histogram(loc,bins=bins_points)
counts=hist[0]
grid_num=hist[1]
grid_num=grid_num[:-1]
```

```python
poro_ns_grid=np.ones((int(Ny),int(Nx)))*(-9999) 
var_grid=np.zeros((int(Ny),int(Nx)))
```

```python
for g in range(0,len(grid_num)):
    index=np.argwhere(loc==grid_num[g])
    ny=np.ceil(grid_num[g]/Nx) 
    nx=(grid_num[g]-(ny-1)*Nx) 
    if index.size !=0: # To avoid division by zero, If the index is different than zero
        poro_ns_grid[int(ny)-1,int(nx)-1]=np.sum(poro_ns[index[:,0],0])/index.shape[0]  # this is the formula for taking the average where there is data
```

```python
unknown_grids=grid_num[counts==0]
```

```python

n_cond_data=12
radius = 4.5
n_real=200
poro_ns_real=np.asmatrix(np.zeros([int(Nx)*int(Ny),n_real])) 
#Loop for realizations
for real in range(0,n_real):
    
    poro_ns_grid_new=poro_ns_grid
    poro_ns_data=poro_ns200 
    X_data=X3M
    Y_data=Y3M
    np.random.shuffle(unknown_grids) 
    print(real)
    for ug in range(0,len(U0_grids)):
        jj = np.ceil(U0_grids[ug]/Nx)
        ii = U0_grids[ug] - (jj - 1) * Nx
        x_cord = ((ii-1)*dx + ii * dx)/2
        y_cord = ((jj-1)*dy + jj * dy)/2
         
    
        dist_U_U0_1 = np.sqrt(np.power(X_data-x_cord,2)+np.power(Y_data-y_cord,2))
        
        points_within_circle = np.argwhere(dist_U_U0_1<radius)
        
        if points_within_circle.shape[0] > n_cond_data:
            points_within_circle = points_within_circle[0:n_cond_data,:]
            
            
        
        #########
        
        ones_vec = np.asmatrix(np.ones(points_within_circle.shape[0]))
        X_mat = X_data[points_within_circle[:,0]]* ones_vec
        Y_mat = Y_data[points_within_circle[:,0]]* ones_vec
        
        
        dist_UU = np.sqrt(np.power(X_mat - X_mat.T,2) + np.power(Y_mat - Y_mat.T,2))
        dist_U_U0 = dist_U_U0_1[points_within_circle[:,0]]
        gamma_UU = sph_var(dist_UU,a,c,nu)
        gamma_UU0 = sph_var(dist_U_U0,a,c,nu)
        C_UU = nu +c- gamma_UU
        C_UU0 = nu + c - gamma_UU0
        cw = np.linalg.inv(C_UU) * C_UU0
        
    
        
        #########
        krig_est=cw.T*(poro_ns_data[points_within_circle[:,0]]) # no need to subtract meanfor simpe Krigin as mean=0 for normal scores
        krig_var=(nu+c)-cw.T*C_UU0
        sample=krig_est+(np.power(krig_var,0.5))*np.random.randn(1) # Extracting a random number from normal scores distribution. Formula below
        poro_ns_grid_new[int(jj)-1,int(ii)-1]=sample
        
        X_data=np.r_[X_data,np.asmatrix(x_cord)] # Appening the coordinates of the new sample to be estimated
        Y_data=np.r_[Y_data,np.asmatrix(y_cord)]
        poro_ns_data=np.r_[poro_ns_data,np.asmatrix(sample)] # Appending the new sample to the data
    
    poro_ns_real[:,real]=poro_ns_grid_new.T.reshape((int(Nx)*int(Ny),1))
 
    
```
