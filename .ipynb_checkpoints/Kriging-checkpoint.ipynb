```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy as sc
from scipy.optimize import curve_fit
import math
from mpl_toolkits.axes_grid1 import make_axes_locatable
from matplotlib.colors import LogNorm
from matplotlib.ticker import MultipleLocator
import matplotlib.image as mpimg
```

```python
Data1=pd.read_csv('poro.txt',delimiter="\t")
Data1.head()
```

```python
East=Data1['% Easting(km)']
North=Data1['Northing(km)']
Poros=Data1['Porosity (%)']
E1=np.asmatrix(East)
N1=np.asmatrix(North).T
Poro1=np.asmatrix(Poros)
n_points1=E1.shape[1]
ones_vec1=np.asmatrix(np.ones(n_points1))
E_mat=E1.T*ones_vec1
N_mat=N1*ones_vec1
Poro_mat=Poro1.T*ones_vec1
```

```python
dist=np.sqrt(np.power((E_mat-E_mat.T),2)+np.power((N_mat-N_mat.T),2))
sqrt_diff=np.power((Poro_mat-Poro_mat.T),2)
sqrt_diff=np.asmatrix(sqrt_diff)
```

## Building Dirctional variogram in NE direction 

```python
lag_size=1
n_lags=21
EDvar=np.zeros(n_lags)
lag_tol=0.5*lag_size
Az=45
angle_tol=45
b_width=7
ual=Az+angle_tol
lal=Az-angle_tol
for l in range(n_lags):
    
    ##compute lag distances
    
    lag=l+1
    ul=lag_size*lag+lag_tol
    ll=lag_size*lag-lag_tol
    index=np.argwhere((dist<=ul)&(dist>=ll))

    ##computing points angles and azimuth
    
    points_ang=np.arctan(np.divide(E_mat[index[:,0]],N_mat[index[:,0]]))*180/np.pi
    ang_id=np.argwhere((points_ang<=ual)&(points_ang>=lal))
    points_ang_af_ang=points_ang[ang_id[:,0]]
    
    ##comparion with band width
    
    index_af_ang=index[ang_id[:,0],:]
    diff_ang=np.abs(Az-points_ang_af_ang)
    Hyp=(np.sqrt(np.power(E_mat[index_af_ang[:,0]],2)+np.power(N_mat[index_af_ang[:,0]],2)))
    dp_w_dv=np.multiply(Hyp,np.sin(diff_ang*np.pi/180))
    id_af_bw=np.argwhere(dp_w_dv<=b_width)
    index_af_bw=index_af_ang[id_af_bw[:,0],:]
    
    
    
    Nh=index_af_bw.shape[0]
    
    EDvar[l]=np.sum(sqrt_diff[index_af_bw[:,0],index_af_bw[:,1]])/(2*Nh)
```

```python
lag_dist_dir=np.arange(lag_size,lag_size*(n_lags+1),lag_size)
plt.figure(figsize=(10,8))
plt.plot(lag_dist_dir,EDvar,'-*' )
plt.title('Dirctional variogram')
plt.grid()
plt.xlabel('dist')
```

```python
def sph_var(dist,rang,c,nugget):
    gamma = nugget+c*(1.5*(dist/rang)-0.5*np.power(dist/rang,3))
    index = np.argwhere(dist>rang)
    gamma[index] = nugget+c
    return gamma
```

```python
def exp_var(dist,rang,sill,nugget):
    if nugget<0:
        nugget = 0
    gamma = nugget+sill*(1-np.exp(-3*dist/rang))
    return gamma
```

```python
def gass_var(dist, rang, c,nugget): 
    gamma=nugget+c*(1-np.exp(-3*np.power(dist/rang,2)))
    index=np.argwhere(dist>rang)
    gamma[index]=nugget+c
    return gamma
```

### Curve fitting: Built-in function

```python
params1,_ = curve_fit(sph_var,lag_dist_dir[:14],EDvar[:14],p0=[3.9,0.5,0.001],bounds=(0,[6,1.5,0.35])) # po:range, sill, nugget ,bounds=(0,[80,20,2.5])
rang1 =  params1[0]
sill1 = params1[1]
nugget1 = params1[2]
print('for Spherical variogram')
print('range is:',round(rang1,4))
print('sill is:',round(sill1,4))
print('nugget effect is:',round(nugget1,4))
gamma_h= sph_var(lag_dist_dir,rang1,sill1,nugget1) # Passing the lag_dist as h
plt.figure(figsize=(10,8))
plt.plot(lag_dist_dir[:14],EDvar[:14],label='Directional_experimatral')
plt.plot(lag_dist_dir[:14],gamma_h[:14],label='Spherical Model') # Fitted function
plt.grid()
plt.legend()
plt.xlabel('dist')
```

```python
params2,_ = curve_fit(gass_var,lag_dist_dir[:14],EDvar[:14],p0=[3.9,0.5,0.001],bounds=(0,[6,1.5,0.35])) # po:range, sill, nugget ,bounds=(0,[80,20,2.5])
rang2 =  params2[0]
sill2 = params2[1]
nugget2 = params2[2]
print('for Gaussian variogram')
print('range is:',round(rang2,4))
print('sill is:',round(sill2,4))
print('nugget effect is:',round(nugget2,4))
gamma_h= gass_var(lag_dist_dir,rang2,sill2,nugget2) # Passing the lag_dist as h
plt.figure(figsize=(10,8))
plt.plot(lag_dist_dir[:14],EDvar[:14],label='Directional_experimatral')
plt.plot(lag_dist_dir[:14],gamma_h[:14],label='Gaussian Model') # Fitted function
plt.grid()
plt.legend()
plt.xlabel('dist')
```

```python
params3,_ = curve_fit(exp_var,lag_dist_dir[:14],EDvar[:14],p0=[3.9,0.5,0.001],bounds=(0,[6,1.5,0.35])) # po:range, sill, nugget ,bounds=(0,[80,20,2.5])
rang3 =  params3[0]
sill3 = params3[1]
nugget3 = params3[2]
print('for Exponential variogram')
print('range is:',round(rang3,4))
print('sill is:',round(sill3,4))
print('nugget effect is:',round(nugget3,4))
gamma_h= gass_var(lag_dist_dir,rang3,sill3,nugget3) # Passing the lag_dist as h
plt.figure(figsize=(10,8))
plt.plot(lag_dist_dir[:14],EDvar[:14],label='Directional_experimatral')
plt.plot(lag_dist_dir[:14],gamma_h[:14],label='Exponential Model') # Fitted function
plt.grid()
plt.legend()
plt.xlabel('dist')
```

# Building grids

```python
X = Data1['% Easting(km)']
Y = Data1['Northing(km)']
Poro = Data1['Porosity (%)']
```

```python
f = plt.figure()
f.set_figwidth(9)
f.set_figheight(6)
plt.scatter(X,Y,c=Poro)
plt.title('Scatter Plot of raw data')
plt.xlabel('X(Km)')
plt.ylabel('Y(Km)')
plt.colorbar()
```

```python
Lx=math.ceil(np.max(X)) #to avoid having float number in Nx, Ny
Ly=math.ceil(np.max(Y))

dx = 0.1      #100m = 0.1 Km
dy = 0.1      #100m = 0.1 Km
Nx = (Lx / dx)
Ny = (Ly / dy) 
```

```python
ix = np.asmatrix(np.ceil(X / dx))
ix[ix==0] = 1
iy = np.asmatrix(np.ceil(Y / dy))
iy[iy==0] = 1
```

```python
f = plt.figure() 
f.set_figwidth(40)
f.set_figheight(32)
plt.scatter(X,Y,c=Poro)
plt.xlabel('X')
plt.ylabel('Y')
plt.xticks(np.arange(0,Lx+dx,dx))
plt.yticks(np.arange(0,Ly+dy,dy))
plt.title('scatter plot of biased data with griding', fontsize=29)
plt.colorbar()
plt.grid()
```

```python
iz=0
grid_id = (iz-1)*Nx*Ny+(iy-1)*Nx+ ix #location of data
```

```python
Poro_grid = np.zeros([int(Nx),int(Ny)]) #building empty matrix 
var_grid = np.zeros([int(Nx),int(Ny)])
```

```python
plt.imshow(Poro_grid,extent=[0,Lx,Ly,0])
plt.xlabel('X')
plt.ylabel('Y')
plt.colorbar()
plt.gca().invert_yaxis()
```

```python
bs = np.arange(1,Nx*Ny+1+1)
output = np.histogram(grid_id,bins=bs)
```

```python
freq = output[0]
bins_num = output[1]
grid_num1 = bins_num[:-1]
```

```python
empt_grid=grid_num1[freq==0] #empty grids
n_grid_ocp=Nx*Ny-len(empt_grid)
```

```python
loc_arr=np.asarray(grid_id) #matrix of locations
```

```python
total_points=Poro.shape[0]
```

### calcuating the weight factor for each grid

```python
loc_arr=np.asarray(grid_id)
wi=np.ones(total_points)
for g in range(0,len(grid_num1)):
    index=np.argwhere(loc_arr==grid_num1[g])
    if index.size !=0:
        wi[index[:,0]]=(total_points/n_grid_ocp)*(1/freq[g])
        
wi        
```

### weight factors for all grids are same, so each data point occupied one grid.

```python
poro_val=Data1['Porosity (%)']
poro_declust=np.multiply(wi,poro_val)
```

```python
 Poro_declust77= np.zeros([int(Nx),int(Ny)])
```

```python
for g in range(0,len(freq)):
    index1 = np.argwhere(grid_id == grid_num1[g])
    j = np.ceil(grid_num1[g]/Nx)
    i = grid_num1[g] - (j-1)*Nx
    if index1.size != 0:
        
        Poro_declust77[int(i)-1,int(j)-1] = np.sum(poro_declust[0,index1[:,1]]) / freq[g]
```

```python
f = plt.figure() 
f.set_figwidth(40)
f.set_figheight(32)
plt.scatter(X,Y,c=poro_declust)
plt.xlabel('X')
plt.ylabel('Y')
plt.xticks(np.arange(0,Lx+dx,dx))
plt.yticks(np.arange(0,Ly+dy,dy))
plt.title('scatter plot of data after declustering', fontsize=29)
plt.colorbar()
plt.grid()
```

```python
U0_grids = grid_num1[freq==0]
```

```python
X_M = np.asmatrix(X).T
Y_M = np.asmatrix(Y).T
poro_declust_M=np.asmatrix(poro_declust).T
```

### Simple Kriging: estimataing the values of porosity and error variance 

```python
for ug in range(0,len(U0_grids)):
    jj = np.ceil(U0_grids[ug]/Nx)
    ii = U0_grids[ug] - (jj - 1) * Nx
    x_cord = ((ii-1)*dx + ii * dx)/2
    y_cord = ((jj-1)*dy + jj * dy)/2
    dist_U_U0_1 = np.sqrt(np.power(X_M - x_cord,2) + np.power(Y_M - y_cord,2))
    points_within_circle = np.argwhere(dist_U_U0_1 < 4.5)
    ones_vec = np.asmatrix(np.ones(points_within_circle.shape[0]))
    X_U = X_M[points_within_circle[:,0]]
    Y_U = Y_M[points_within_circle[:,0]]
    X_mat = X_U * ones_vec
    Y_mat = Y_U * ones_vec
    dist_UU = np.sqrt(np.power(X_mat - X_mat.T,2) + np.power(Y_mat - Y_mat.T,2))
    dist_U_U0 = dist_U_U0_1[points_within_circle[:,0]]
    gamma_UU = exp_var(dist_UU,a,c,nu)
    gamma_UU0 = exp_var(dist_U_U0,a,c,nu)
    C_UU = nu +c- gamma_UU
    C_UU0 = nu + c - gamma_UU0
    cw = np.linalg.inv(C_UU) * C_UU0
    Poro_grid[int(ii)-1,int(jj)-1] =cw.T * ( poro_declust_M[points_within_circle[:,0]] - np.mean(poro_declust_M)) + np.mean(poro_declust_M)
    var_grid[int(ii)-1,int(jj)-1] = (nu + c) - (cw.T * C_UU0)
   
```

```python
f = plt.figure() 
f.set_figwidth(40)
f.set_figheight(32)
plt.imshow(Poro_grid.T,extent=[0,Lx,0,Ly],cmap='jet')
plt.title('Simple Kriging', fontsize=29)
plt.xlabel('X(Km)', fontsize=15)
plt.ylabel('Y(Km)', fontsize=15)
plt.gca().invert_yaxis()
plt.colorbar()
```

```python
Poro_grid_ok = np.zeros([int(Nx),int(Ny)])
var_grid_ok = np.zeros([int(Nx),int(Ny)])
```

### Ordinary Kriging: estimataing the values of porosity and error variance 

```python
for ug in range(0,len(U0_grids)):
    jj = np.ceil(U0_grids[ug]/Nx)
    ii = U0_grids[ug] - (jj - 1) * Nx
    x_cord = ((ii-1)*dx + ii * dx)/2
    y_cord = ((jj-1)*dy + jj * dy)/2
    dist_U_U0_1 = np.sqrt(np.power(X_M - x_cord,2) + np.power(Y_M - y_cord,2))
    points_within_circle = np.argwhere(dist_U_U0_1 < 4.5)
    ones_vec = np.asmatrix(np.ones(points_within_circle.shape[0]))
    X_U = X_M[points_within_circle[:,0]]
    Y_U = Y_M[points_within_circle[:,0]]
    X_mat = X_U * ones_vec
    Y_mat = Y_U * ones_vec
    dist_UU = np.sqrt(np.power(X_mat - X_mat.T,2) + np.power(Y_mat - Y_mat.T,2))
    dist_U_U0 = dist_U_U0_1[points_within_circle[:,0]]
    gamma_UU = exp_var(dist_UU,a,c,nu)
    gamma_UU0 = exp_var(dist_U_U0,a,c,nu)
    C_UU = nu +c- gamma_UU
    C_UU0 = nu + c - gamma_UU0
    ###
    
    COK=np.r_[C_UU,ones_vec]  ##ones__
    col_vec=np.r_[ones_vec.T,np.asmatrix(0)]
    c_ok=np.c_[COK,col_vec]
    
    
    co_ok=np.r_[C_UU0,np.asmatrix(1)]
               
                   
              
    cw_ok = np.linalg.inv(c_ok) * co_ok
    
    
    
    Poro_grid_ok[int(ii)-1,int(jj)-1] =cw_ok[:-1].T * poro_declust_M[points_within_circle[:,0],0]
    var_grid_ok[int(ii)-1,int(jj)-1] = (nu+c)  - cw_ok[:-1].T * co_ok[:-1] - cw_ok[-1]
    
```

```python
f = plt.figure() 
f.set_figwidth(40)
f.set_figheight(32)
plt.imshow(Poro_grid_ok.T,extent=[0,Lx,Ly,0],cmap='jet')
plt.title('Ordinary Kriging', fontsize=29)
plt.xlabel('X(Km)', fontsize=15)
plt.ylabel('Y(Km)', fontsize=15)
plt.gca().invert_yaxis()
plt.colorbar()
```
